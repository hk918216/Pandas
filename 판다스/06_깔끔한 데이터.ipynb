- 하나의 열이 여러 의미 가지는 경우

```python
# split(), concat()
ebola = pd.read_csv('C:/STUDY/Pandas/country_timeseries.csv')
ebola.iloc[:5,[0,1,2,3,10,11]]
```

```python
ebola_long = pd.melt(ebola, id_vars=['Date','Day'])
ebola_long
```

```python
# vairable 열 데이터를 _ 기준으로 분리
variable_split = ebola_long.variable.str.split('_')  # ebola_long['variable'].str.split('_')
variable_split[:5]
```

```python
# 분리된 데이터 확인
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)

print(status_values[:5])
print(country_values[:5])
```

```python
# 분리된 데이터를 ebola_long 데이터프레임에 추가
ebola_long['ststus'] = status_values
ebola_long['country'] = country_values
ebola_long.head()
```

```python
# concat()으로 합치기
variable_split = ebola_long.variable.str.split('_',expand=True)
variable_split.columns = ['status','country']
ebola_parsed = pd.concat([ebola_long, variable_split], axis = 1)

ebola_parsed.head()
```

```python
ebola_parsed.loc[:,['Date','Day','status','country','value']]
```

#### 기상 데이터의 여러 열을 하나로 정리하기

```python
weather = pd.read_csv('C:/STUDY/Pandas/weather.csv')
print(weather.shape)
print(weather.columns)
print(weather.iloc[:5,:11])
```

```python
weather_melt = pd.melt(weather,
                      id_vars=['id','year','month','element'],
                      var_name='day', value_name='temp')

weather_melt[100:110]
```

```python
# pivot_table()사용
weather_tidy = weather_melt.pivot_table(
                index=['id','year','month','day'],
                columns = 'element', values='temp')

weather_tidy.head(10)
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat.head()
```

#### 중복데이터 처리

```python
billboard.head()
```

```python
billboard_long = pd.melt(billboard,
                         id_vars=['year','artist','track','time','date.entered'],
                         var_name='week',value_name='rating')
print(billboard_long.shape)
print(billboard_long.head(10))
```

```python
billboard_long[billboard_long.track == 'Loser'].head()
```

```python
billboard_song = billboard_long[['year','artist','track','time']]
print(billboard_song)
```

```python
# 중복데이터 제거
billboard_song = billboard_song.drop_duplicates()
print(billboard_song.shape)
```

```python
# 순번 추가
billboard_song['id'] = range(len(billboard_song))
billboard_song.head(10)
```

```python
# 노래 정보와 주간 순위 데이터 합치기
billboard_ratings = billboard_long.merge(
    billboard_song, on=['year','artist','track','time'])

print(billboard_ratings.shape)
billboard_ratings.head()
```
