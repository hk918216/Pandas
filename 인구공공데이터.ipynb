#### 성별 / 연령별

```python
df = pd.read_csv('C:/STUDY/people_gender_202101.csv', encoding='cp949', thousands=',')
df.head()
```

```python
df2 = df[df['행정구역'] == '부산광역시 사하구 괴정제4동(2638054000)']
print(df2)
```

```python
df2 = df2.drop(['행정구역','2021년01월_남_총인구수','2021년01월_남_연령구간인구수',
               '2021년01월_여_총인구수','2021년01월_여_연령구간인구수'], axis=1)
df2.head()
```

```python
# 남자데이터 확인
male = df2.iloc[:,0:11]
male
```

```python
# 여자데이터 확인
female = df2.iloc[:,11:22]
female
```

```python
# 남자데이터정리
male2 = male.melt()
male2.drop('variable', axis=1, inplace=True)
male2['value'] = male2['value'].apply(lambda x : x * -1) # = male2= male2 * -1 
male2
```

```python
# 여자데이터정리
female2 = female.melt()
female2.drop('variable', axis=1, inplace=True)

female2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title('부산 사하구 남녀 성별 인구 분포')
plt.barh(male2.index, male2['value'].tolist(), label='남자')
plt.barh(female2.index, female2['value'].tolist(), label='여자')

plt.yticks(range(11), ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80-89','90-99','100이상'])
plt.xlim(-4500,4500)
plt.xticks(range(-4000,4001,2000), ['4000명','2000명','0명','2000명','4000명'])

plt.legend()
plt.show()
```
